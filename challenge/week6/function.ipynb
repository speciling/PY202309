```python
import random

# 1~45 사이의 랜덤한 숫자 한 개 반환
def get_number():
    return random.randrange(1, 46)

# 1~45 사이의 중복없는 숫자 6개의 리스트 반환
def set_lotto_numbers():
    lotto = []
    # lotto 리스트의 갯수가 6보다 작으면 반복(6개가 되면 종료)
    while len(lotto) < 6:
        # 랜덤 숫자 num 변수에 저장
        num = get_number()
        # num이 lotto 리스트 안에 없으면 lotto 리스트에 num 추가
        if num not in lotto:
            lotto.append(num)
    # lotto 리스트 반환
    return lotto

# 6개의 숫자가 든 두 리스트의 중복된 원소를 모아둔 집합 반환
def match_numbers(user_inputs, lotto_numbers):
    # 각 리스트를 set()함수를 통해 집합으로 변환
    # & 연산자로 교집합을 구해 반환
    return set(user_inputs) & set(lotto_numbers)

# 같은 숫자들의 갯수에 따른 등수 출력(반환값 없음)
def check_prize(same_numbers):
    # len함수를 사용해 같은 숫자의 갯수를 구해 조건문으로 갯수에 따른 등수 출력
    if len(same_numbers) == 6:
        print("1st prize")
    elif len(same_numbers) == 5:
        print("2nd prize")
    elif len(same_numbers) == 4:
        print("3rd prize")
    elif len(same_numbers) == 3:
        print("4th prize")
    else:
        print("no prize :(")

num = 0
print("** Start Lotto **")
print("1) Set the lotto numbers")
# 정답 로또 번호를 생성하여 lotto_numbers에 저장
lotto_numbers = set_lotto_numbers()

print("2) Set the user inputs")
user_inputs = []
# user_inputs가 6보다 작으면 반복(6개가 되면 종료)
while len(user_inputs) < 6:
    # 입력받은 수를 num에 저장
    num = int(input("Enter %d-number: "%(len(user_inputs)+1)))
    # 1~45 바깥의 수면 No valid number 출력
    if num < 1 or num > 45:
        print("No valid number")
    # user_inputs 안에 있으면(이미 입력받은 수라면) Same number 출력
    elif num in user_inputs:
        print("Same number")
    # 범위 안의 입력받은적 없는 수라면 user_inputs에 추가
    else:
        user_inputs.append(num)
        
print("3) Match the numbers")
# 정답 로또 번호와 사용자 입력 로또 번호중 일치하는 숫자들을 모은 집합 저장
same_numbers = match_numbers(user_inputs, lotto_numbers)

print("4) Check the prize")
# 일치하는 수들의 갯수에 따른 등수 출력
check_prize(same_numbers)

print("5) Sort the lists")
# 로또 번호 리스트들을 오름차순으로 정렬
lotto_numbers.sort()
user_inputs.sort()

print("=== Summary ===")
# *를 사용하여 각 리스트의 모든 원소를 출력
print("Lotto numbers: ", *lotto_numbers)
print("User numbers: ", *user_inputs)
```
